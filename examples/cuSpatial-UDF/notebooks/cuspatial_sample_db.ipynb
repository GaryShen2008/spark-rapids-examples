```
#define the input & output path
inputPath='dbfs:/data/cuspatial_data/points'
outputPath='dbfs:/data/output'

# add the shapefile(.shp & .shx)
spark.sparkContext.addFile("dbfs:/data/cuspatial_data/polygons/polygons.shp")
spark.sparkContext.addFile("dbfs:/data/cuspatial_data/polygons/polygons.shx")
```

```
# register the UDF
spark.udf.registerJavaFunction("point_in_polygon", "com.nvidia.spark.rapids.udf.PointInPolygon", None)

df = spark.read.parquet(inputPath)
df = df.selectExpr('x', 'y', 'point_in_polygon(x, y) as ret')
df.write.mode("overwrite").parquet(outputPath)
```

```
print("Input rows: ", df.count())
```

```
# show the result
result = spark.read.parquet(outputPath)
print("Output rows: ", result.count())
result.show(20)
```
